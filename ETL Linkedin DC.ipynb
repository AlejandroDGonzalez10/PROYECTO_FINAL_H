### Obtención de datos, transformación y selección de los mismos para futuro analisís. 
importamos las librerias necesarias.

```python
import pandas as pd # Importar las bibliotecas necesarias
import numpy as np
import pycountry
import re
```

el siguiente codigo llama al csv selecionado

```python
#Leer el archivo CSV
DS_df = pd.read_csv('PF/linkedin-jobs-Latam_DS.csv', delimiter=',', encoding='utf-8')
DS_df
```

verificamos que no haya duplicados

```python
DS_df.duplicated().sum() #contar el número de filas duplicadas
```

lo mismo con los nulos


```python
DS_df.isnull().sum() #contar el número de valores faltantes
```

si fuera el caso procedemos a eliminar los duplicados y elimino la primera fila que tiene datos duplicados de las columnas

```python
DS_df = DS_df.drop_duplicates() #eliminar filas duplicadas 
DS_df.drop(DS_df.index[0], inplace=True)
```

```python
DS_df # mostrar las primeras filas de un objeto DataFrame
```

contamos los valores que aparecen en la columna criteria.

```python
DS_df.value_counts('criteria') #contar el número de veces que aparece cada valor en la columna
```

recategorizamos las filas de la columna criteria en nivel de experienciay a tipo de empleo tambien lo pasamos al castellano


```python
DS_df["nivel_de_experiencia"] = DS_df["criteria"].str.extract(r"'(Nível de experiência|Nivel de antigüedad)':\s+'([^']+)'\s*}")[1]
#crear una nueva columna en un objeto DataFrame
DS_df["tipo_empleo"] = DS_df["criteria"].str.extract(r"'(Tipo de emprego|Tipo de empleo)':\s+'([^']+)'\s*}")[1]

DS_df
```

contamos los valores de nivel de experiencia y tipo de empleo

```python
DS_df.value_counts('nivel_de_experiencia') # contar los valores únicos de la columna
```

```python
DS_df.value_counts('tipo_empleo') #contar los valores únicos de la columna
```

recategorizamos los valores de la filas.

```python
#reemplazar los valores en la columna "nivel_de_experiencia" de un objeto DataFrame llamado DE_df, de acuerdo con un diccionario de valores a reemplazar.
DS_df['nivel_de_experiencia'] = DS_df['nivel_de_experiencia'].replace({'Não aplicável': np.nan,
                                                                       'Pleno-sênior': 'Semi-senior-Senior',
                                                                       'Assistente': 'Ayudante',
                                                                       'Estágio': 'Pasante',
                                                                       'Júnior': 'Junio'})

DS_df
```

recategorizamos los valores de las filas de tipo de empleo


```python
#reemplazar los valores en la columna "nivel_de_experiencia" de un objeto DataFrame llamado DE_df, de acuerdo con un diccionario de valores a reemplazar.
DS_df['tipo_empleo'] = DS_df['tipo_empleo'].replace({'Tempo integral': 'Fulltime',
                                                                       'Jornada completa': 'Fulltime',
                                                                       'Media jornada': 'Part time',
                                                                       'Contrato': 'Otro',
                                                                       'Outro': 'Otro'})

DS_df
```

eliminamos las columnas que no nos parece relevantes para un futuro analisis

```python
DS_df = DS_df.drop(['criteria','link'], axis=1)#eliminar las columnas "localidad" y "requisitos" del objeto DataFrame

DS_df
```

cambiar el valor de los NAN

```python
DS_df.fillna(0, inplace=True) #cambiar las casillas con valores faltantes
```

renombramos las columnas a nombres que nos parece correctos

```python
DS_df = DS_df.rename(columns={'title': 'titulo_puesto', 'company': 'empresa', 'description': 'descripcion_puesto', 'onsite_remote':'modalidad','salary':'salario','posted_date':'fecha_posteo','location':'localidad'})
```

usamos la libreria pycountry para asociar los paises en la columna localidad

```python
localidades_unicas = DS_df['localidad'].unique()

paises = [pais.name for pais in pycountry.countries]
def identificar_pais(cadena):
    for pais in paises:
        if pais in cadena:
            return pais
    return cadena

DS_df['localidad'] = DS_df['localidad'].apply(identificar_pais)
```

terminamos por convertir el archivo a csv



```python
#Guardar un objeto DataFrame llamado DE_df en un archivo CSV 
DS_df.to_csv('linkedin_DS_empleos.csv', index=False, encoding='utf-8')
```
